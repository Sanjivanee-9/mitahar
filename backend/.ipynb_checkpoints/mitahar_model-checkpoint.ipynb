```python
import pandas as pd

data = pd.read_csv('recipes.csv')
data.info()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 8))
plt.title('No.of Food Items vs Calories')
plt.ylabel('No. of Food Items')
plt.xlabel('Calories')
ax.hist(data.Calories.to_numpy(),bins=[0,100,200,300,400,500,600,700,800,900,1000,2000,3000,4000,5000],linewidth=0.5, edgecolor="white")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
plt.title('No.of Food Items vs FatContent')
plt.ylabel('No. of Food Items')
plt.xlabel('FatContent')
ax.hist(data.Calories.to_numpy(),bins=[0,10,20,30,40,50],linewidth=0.9, edgecolor="white")
plt.show()
```
